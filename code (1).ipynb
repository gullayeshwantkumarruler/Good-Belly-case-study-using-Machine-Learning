```python
import pandas as pd

df = pd.read_excel(r'/content/drive/MyDrive/Freelance/Alberto/belly_data/GoodBelly_dataset.xlsx', sheet_name='Real Data')
df.head()
```

```python
data=df
```

```python
import statsmodels.api as sm

X = data[["Average Retail Price", "Sales Rep", "Endcap", "Demo", "Demo1-3", "Demo4-5", "Natural", "Fitness"]]

# add intercept term
X = sm.add_constant(X)

# define response variable
y = data["Units Sold"]

```

```python
# fit multiple linear regression model
model = sm.OLS(y, X).fit()

# print summary of regression output
print(model.summary())
```

## Question-5
<b>   Are there any suggestions to improve and refine the model?

<b>Solution</b><br>
There are several suggestions to improve and refine the above model:

Include additional variables: There may be other variables that could have an impact on GoodBelly's sales that were not included in the current model. For example, variables such as weather conditions or seasonality may affect the sales of GoodBelly's products.

Test for non-linear relationships: The current model assumes that the relationship between each independent variable and the dependent variable is linear. However, there may be non-linear relationships between the variables that could be better captured by a more flexible model.

Check for multicollinearity: Multicollinearity occurs when independent variables in the model are highly correlated. This can lead to unstable coefficients and make it difficult to interpret the results. Therefore, it is important to check for multicollinearity and consider removing one of the highly correlated variables if it exists.

Consider using a different regression method: The current model uses ordinary least squares (OLS) regression. However, other regression methods such as robust regression or ridge regression may be more appropriate for the current data set depending on the distribution of the data and the presence of outliers.

Explore interactions between variables: The current model does not include any interaction terms, which could help capture more complex relationships between the variables. For example, the effect of demos may vary depending on whether a store had a regional sales rep or not.

Check for outliers: Outliers can have a significant impact on the coefficients and the overall fit of the model. Therefore, it is important to identify and investigate any outliers and consider removing them if they are influential.
